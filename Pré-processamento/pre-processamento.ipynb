**Sobre o conjunto de dados**

Este é um conjunto de dados que requer muito pré-processamento com insights EDA incríveis para uma empresa. Um conjunto de dados que consiste em dados de vendas e lucros classificados por segmento de mercado e país/região.

Importando a base de dados

```python
# Importando bibliotecas
import pandas as pd

# Configurandoo pandas para exibir todas as colunas
pd.set_option('display.max_columns', None)
```

```python
# Importando a base de dados

df = pd.read_csv('Financials.csv', encoding='ISO-8859-1')

display(df)
```

```python
# Analisando os tipos dos dados
empresa = df.info()

print(empresa)
```

```python
# Renomeando as colunas

df = df.rename(columns={
    'Segment': 'segmento',
    'Country': 'país',
    ' Product ': 'produto',
    ' Discount Band ': 'faixa_de_desconto',
    ' Units Sold ': 'unidades_vendidas',
    ' Manufacturing Price ': 'preço_de_fabricação',
    ' Sale Price ': 'preço_de_venda',
    ' Gross Sales ': 'vendas_brutas',
    ' Discounts ': 'descontos',
    '  Sales ': 'vendas',
    ' COGS ': 'cogs',
    ' Profit ': 'lucro',
    'Date': 'data',
    'Month Number': 'número_do_mês',
    ' Month Name ': 'nome_do_mês',
    'Year': 'ano'
})
```

```python
# Removendo a coluna numero do mês

display(df.head(20))
```

```python
# Removendo caracteres especiais e convertendo o tipo das variavéis qualitativas
variaveis_limpeza = ['unidades_vendidas', 'preço_de_fabricação', 'preço_de_venda', 'vendas_brutas', 'descontos', 'vendas', 'cogs', 'lucro']

for var in variaveis_limpeza:
    df[var] = df[var].replace('[\$,()\-\,]', '', regex=True)
    df[var] = pd.to_numeric(df[var], errors='coerce')
```

```python
# Verificando o tipo dos dados
empresa = df.info()

print(empresa)
```

```python
valores = df.isna().sum()
print(valores)
```

```python
# Substituindo os valores Nan por 0
df['descontos'] = df['descontos'].fillna(0)

df['lucro'] = df['lucro'].fillna(0)
```

```python
display(df)
```

```python
# Convertendo as variavéis data e ano

df['data'] = pd.to_datetime(df['data'])

df['ano'] = pd.to_datetime(df['ano'], format='%Y')
```

```python
# Removendo a variavél
df = df.drop(columns=['número_do_mês'])
```

```python
# Visualizando o tipo dos dados

df.info()
```

```python
# Exportando a base de dados tratada
df.to_csv('dados_tratados.csv', index=False)
```
